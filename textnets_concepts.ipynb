```python
import textnets as tn
from concepts import Context
```

```python
tn.params["autodownload"] = True
tn.params["seed"] = 100
```

```python
n = tn.Textnet(tn.Corpus(tn.examples.digitalisierung, lang="de").noun_phrases())
```

```python
ctx = Context(*n.context)
```

```python
v = ctx.lattice.graphviz(make_property_label="\n".join)
```

```python
v.edge_attr.update(fontsize="6")
```
