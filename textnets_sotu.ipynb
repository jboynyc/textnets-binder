```python
%matplotlib inline
```

```python
import textnets as tn
```

```python
tn.params["seed"] = 23
```

```python
#!wget https://github.com/tm4ss/tm4ss.github.io/raw/master/docs/data/sotu.csv
```

```python
corpus = tn.Corpus.from_csv("sotu.csv", sep=";", 
                            doc_col="text", 
                            label_col="president")
```

```python
np = corpus.noun_phrases()
```

```python
n = tn.Textnet(np, remove_weak_edges=True)
```

```python
presidents = n.project(node_type=tn.DOC)
```

```python
presidents.plot(
    alpha=.125,
    label_nodes=True, 
    node_label_size=6, 
    scale_nodes_by="degree",
    scale_edges_by="weight",
    color_clusters=True,
)
```
